The above boxplots shows the presence of outliers in some features. We need to remove those outliers with the help of zscore and IQR method.

```python
from scipy.stats import zscore
outliers=df[['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Loan_Status']]
z=np.abs(zscore(outliers))
z
```

```python
np.where(z>3)
```

```python
z.iloc[94,9]
```

```python
df1=df[(z<3).all(axis=1)]
df1
```

```python
print('Old data shape',df.shape)
print('New data shape',df1.shape)
```

```python
print('Data loss percentage',((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

The above mentioned data loss is absolutely affordable.

```python
Q1=df1.quantile(.25)
Q2=df1.quantile(.75)
IQR=Q2-Q1
IQR
```

```python
df1_new=df1[(z<3).all(axis=1)]
df1_new
```

```python
plt.figure(figsize=(25,20))
p=1
for i in outliers:
    if p<=18:
        ax=plt.subplot(5,4,p)
        sns.distplot(df[i],color='r')
        plt.xlabel(i,fontsize=15)
    p+=1
plt.show()
```

Here, we have cleared all the outliers from the data using zscore and IQR method. We can see from the above plots, all the features are showing almost normal distribution.
